# Problem 3

the breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information;

```python
mu=5
sigma=1.5

# import the required function
from scipy.stats import norm
```

CDF: of a Random variable ( X ) is the probability that X will take the value less than or equal to x. It can be represented mathematically as -

Fx(x)=P(X≤x)

The library scipy.stats provides a function called cdf() to calculate the CDF of any distribution. Here, we assumed our data to be normal distribution so we can use norm.cdf() to calculate cumulative probability density for a normal distribution

```python
# plot
rand_number = np.arange(-50,120,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
```

**3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?**

```python
prob_less_3 = norm.cdf(3.17, mu, sigma)
print('The probability that the gunny bags have a breaking strength less than 3.17 kg per sq cm',prob_less_3)
```

```python
plt.rcParams["figure.figsize"] = (10,6.5)
# plot
rand_number = np.arange(-50,120,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
plt.axvline(x=3.17, c="r") # draw a red vertical line
x1 = np.linspace(3.17, rand_number.min(), 120)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="r")
```

**3.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?**

```python
prob_less_36 = 1-norm.cdf(3.6, mu, sigma)
print('The probability that the gunny bags have a breaking strength at least 3.6 kg per sq cm',prob_less_36)
```

```python
# plot
rand_number = np.arange(-50,120,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
plt.axvline(x=3.6, c="r") # draw a red vertical line
x1 = np.linspace(3.6, rand_number.max(), 120)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="r")
```

**3.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?**

```python
prob_5 = norm.cdf(5, mu, sigma)
prob_55 = norm.cdf(5.5, mu, sigma)
print('The probability that the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.',prob_55 - prob_5)
```

```python
# plot
rand_number = np.arange(-50,50,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
plt.axvline(x=5, c="r") # draw a red vertical line
plt.axvline(x=5.5, c="r") # draw a red vertical line
```

**3.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?**

```python
prob_3 = norm.cdf(3, mu, sigma)
prob_75 = norm.cdf(7.5, mu, sigma)
prob_no = prob_75 - prob_3
print('The probability that the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.',1 - prob_no)
```

**4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?**

```python
prob_less_85 = norm.cdf(85, mu, sigma)
print('The probability that a randomly chosen student gets a grade below 85 on this exam is',prob_less_85)
# plot
rand_number = np.arange(0,150,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
plt.axvline(x=85, c="r") # draw a red vertical line
x1 = np.linspace(85, rand_number.min(), 150)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="r")
```

**4.2 What is the probability that a randomly selected student scores between 65 and 87?**

```python
prob_87 = norm.cdf(87, mu, sigma)
prob_65 = norm.cdf(65, mu, sigma)
print('The probability that a randomly selected student scores between 65 and 87 is',prob_87 - prob_65)
# plot
rand_number = np.arange(-50,200,0.001)
plt.plot(rand_number,norm.pdf(rand_number,mu,sigma))
plt.axvline(x=87, c="r") # draw a red vertical line
plt.axvline(x=65, c="r") # draw a red vertical line
x1 = np.linspace(87, rand_number.min(), 200)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="r")
x1 = np.linspace(65, rand_number.min(), 200)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color="w")
```

Step 1: define null and alternative hypotheses                                                                                  
null hypotheses: μ1 ≥ 150                                                                                                        
alternative hypotheses: μ1 < 150

Step2: decide the significance level

```python
print('The sample size for this problem is',len(df_z))
```

Step3: Identify the test statistic                                                                                               
as standard deviation is unknown and n=75. so we use the t distribuction and t statistic

Step4: Calculate the p-value and test statistic

```python
# null sample t-test
# null nypothesis: expected value=5
t_statistic ,p_value = ttest_1samp(df_z, 150)
print('One sample t_test \nt statistic: {0} p_value: {1} '.format(t_statistic,p_value))
```

```python
unpo_t_stats=ttest_1samp(df_z, 150)[0][0]
po_t_stats=ttest_1samp(df_z, 150)[0][1]
unpo_p_value=ttest_1samp(df_z, 150)[1][0]
po_p_value=ttest_1samp(df_z, 150)[1][1]
```

Step5 : Decide to reject or accept null hypothesis

```python
# p_value < 0.05 => alternative hypothesis:
α = 0.05
print('level of significance:',α)
print('p-value:' ,unpo_p_value,'\n')
if unpo_p_value < α:
    print('We have evidence to reject the null hypothese since the p-value < level of significance')
else:
    print('We have no evidence to reject the null hypotheses since the p-value > level of significance')
```

Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%)

Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly
